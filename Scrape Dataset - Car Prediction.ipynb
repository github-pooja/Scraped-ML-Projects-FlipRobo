```python
#car_details elements
Brand=[]
Manu_year=[] 
car_model=[]
Variant=[]
Price=[]
Location=[]
Driven_km=[]
Fuel_Type=[]

try:
    Brand_elements=driver.find_elements_by_xpath("//div[@class='gsc_col-xs-7 carsName']/a")
    for q in Brand_elements:
        Brand.append(q.text.strip().split(' ')[1].strip())
except NoSuchElementException:
    Brand.append("--")
    
#Scrapping Manufacturing year
try:
    for m in Brand_elements:
        Manu_year.append(m.text.strip().split(' ')[0].strip())
except NoSuchElementException:
        Manu_year.append("--")
#Manu_year

#Scrapping Car model
try:
    for n in Brand_elements:
        car_model.append(n.text.strip().split(' ')[2].strip())
except NoSuchElementException:
        car_model.append("--")
#car_model

#Scrapping Car Variant
try:
    variant_elements=driver.find_elements_by_xpath("//div[@style='min-height: 16px; color: rgba(0, 0, 0, 0.6);']")
    for o in variant_elements:
        Variant.append(o.text)
except NoSuchElementException:
        Variant.append("--")
#Variant

#Scrapping Price of used cars
try:
    price_elements=driver.find_elements_by_xpath("//span[@class='amnt ']")
    for p in price_elements:
        Price.append(p.text)
except NoSuchElementException:
        Price.append("--")
#Price

#Scrapping Location of used cars
try:
    car_location_elements=driver.find_elements_by_xpath("//div[@class='cityIcon']")
    for r in car_location_elements:
        Location.append(r.text)
except NoSuchElementException:
        Location.append("--")

#Location

#Scrapping Driven KM of used cars
try:
    driven_km_elements=driver.find_elements_by_xpath("//div[@class='truncate dotlist']")
    for s in driven_km_elements:
        Driven_km.append(s.text.strip().split(' ')[0].strip())
except NoSuchElementException:
        Driven_km.append("--")


#Scrapping Fuel of used cars
try:
    for t in driven_km_elements:
        Fuel_Type.append(t.text.strip().split('kms')[1].strip())
except NoSuchElementException:
        Fuel_Type.append("--")
#Fuel

    

```

```python
#car_details elements
Brand3=[]
Manu_year3=[] 
car_model3=[]
Variant3=[]
Price3=[]
Location3=[]
Driven_km3=[]
Fuel_Type3=[]

try:
    Brand_elements3=driver.find_elements_by_xpath("//div[@class='gsc_col-xs-7 carsName']/a")
    for q in Brand_elements3:
        Brand3.append(q.text.strip().split(' ')[1].strip())
except NoSuchElementException:
    Brand3.append("--")
    
#Scrapping Manufacturing year
try:
    for m in Brand_elements3:
        Manu_year3.append(m.text.strip().split(' ')[0].strip())
except NoSuchElementException:
        Manu_year3.append("--")
#Manu_year

#Scrapping Car model
try:
    for n in Brand_elements3:
        car_model3.append(n.text.strip().split(' ')[2].strip())
except NoSuchElementException:
        car_model3.append("--")
#car_model

#Scrapping Car Variant
try:
    variant_elements3=driver.find_elements_by_xpath("//div[@style='min-height: 16px; color: rgba(0, 0, 0, 0.6);']")
    for o in variant_elements3:
        Variant3.append(o.text)
except NoSuchElementException:
        Variant3.append("--")
#Variant

#Scrapping Price of used cars
try:
    price_elements3=driver.find_elements_by_xpath("//span[@class='amnt ']")
    for p in price_elements3:
        Price3.append(p.text)
except NoSuchElementException:
        Price3.append("--")
#Price

#Scrapping Location of used cars
try:
    car_location_elements3=driver.find_elements_by_xpath("//div[@class='cityIcon']")
    for r in car_location_elements3:
        Location3.append(r.text)
except NoSuchElementException:
        Location3.append("--")

#Location

#Scrapping Driven KM of used cars
try:
    driven_km_elements3=driver.find_elements_by_xpath("//div[@class='truncate dotlist']")
    for s in driven_km_elements3:
        Driven_km3.append(s.text.strip().split(' ')[0].strip())
except NoSuchElementException:
        Driven_km3.append("--")


#Scrapping Fuel of used cars
try:
    for t in driven_km_elements3:
        Fuel_Type3.append(t.text.strip().split('kms')[1].strip())
except NoSuchElementException:
        Fuel_Type3.append("--")
#Fuel

    

```
